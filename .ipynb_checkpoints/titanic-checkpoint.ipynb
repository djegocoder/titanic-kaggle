**Criando funções de Análise Exploratória de Dados** 

```python
def resume(df):
    """
    Objective: For a given dataframe this function provides information
    regarding Missing and Unique values per column.

    Input: param df: Dataframe to check the information.

    Output: return summary: a dataframe with columns providing summary per column of the input dataframe.
    
    """
    df = df.copy()
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name','dtypes']]
    summary['Missing'] = df.isnull().sum().values
    summary['Missing Percentage'] = df.isnull().sum().values/len(df)
    summary['Uniques'] = df.nunique().values
    return summary
```

```python
def feature_eng(dataframe):
    """
    Feature engineering process in the data 
    """
    dataframe = dataframe.copy()

    #Ticket Number and Ticket Code features creation spliting 
    # the Letters and Numbers of Ticket feature
    Ticket = dataframe['Ticket'].str.split(" ",expand=True)
    for i in range(dataframe.shape[0]):
        if Ticket.loc[i,2] != None:
            Ticket.loc[i,0] = Ticket.loc[i,0]+'_'+Ticket.loc[i,1]#Sometimes the ticket letter has 2 letters groups 
            Ticket.loc[i,1] = Ticket.loc[i,2]
            Ticket.loc[i,2] = None      
        if Ticket.loc[i,1] == None and Ticket.loc[i,0] != 'LINE':# Line is the unique code without number
            Ticket.loc[i,1] = Ticket.loc[i,0]
            Ticket.loc[i,0] = None     
        if Ticket.loc[i,0]== 'LINE':
            Ticket.loc[i,1] = 0
    dataframe["ticketCod"] = Ticket[0]
    dataframe['ticketNum'] = Ticket[1]
    dataframe['ticketNum'] = dataframe['ticketNum'].astype(int)
    dataframe.drop(['Ticket'],axis=1)

    #Numbers of Cabins and Cabins Letter Code features creation
    dataframe['Cabin'] = dataframe['Cabin'].fillna('Z00')
    Cabin = dataframe["Cabin"].str.split(" ",n=3,expand=True)
    CabinL = Cabin[0]
    CabinL = CabinL.str.split("",n=2,expand=True)
    for i in range(dataframe.shape[0]):
        if Cabin.loc[i,2] != None:
            Cabin.loc[i,3] = 3
        else:
            if Cabin.loc[i,1] != None:
                Cabin.loc[i,3] = 2
            else:
                Cabin.loc[i,3] = 1
        CabinL.loc[i,0] = CabinL.loc[i,1]
    dataframe['cabinLetter'] = CabinL[0]
    dataframe['cabins'] = Cabin[3]
    dataframe['cabins'] = dataframe['cabins'].astype(int)


    #Split Name, Last Name and Treatment Pronoun
    LastName = dataframe["Name"].str.split(",",n=1,expand=True)
    TreatPron = LastName[1].str.split(".",n=1,expand=True)
    dataframe['treatPron'] = TreatPron[0]
    dataframe["firstNam"] = TreatPron[1]
    dataframe["lastNam"] = LastName[0]

    #Invert order of Passenger Class number
    dataframe['Pclass'] = dataframe['Pclass'].replace([1,3],[3,1])
        
    return dataframe
```
